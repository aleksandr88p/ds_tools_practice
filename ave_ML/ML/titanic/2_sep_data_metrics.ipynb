    В машиином обучении обучаем комп делать выводы на основе данных
    что бы не было переобучения надо давать другие данные на тесте. 
    значит надо разделить датасет на три части:
     - на тренировочные(на которых обучаем)
     - валидационный набор (те которые модель не видела) что бы подбирать гиперпараметры
     - тестовый набор (для оценки того как модель работает)

    Норм прапорция 60/20/20 (train, val, test)

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split

titanic_csv = 'train.csv'

titanic = pd.read_csv(titanic_csv)  
```

```python
titanic.head()
```

```python
# перед тем как делать данные на тренировачные и тестовые, нужно определить их признаки или лейблы(метки, target)
# train, val, test sets split

# то на чем будем учиться
feutures =  titanic.drop('Survived', axis=1)


# то что будем предсказывать
labels = titanic['Survived']

# делаю на этом шаге 40 процентов для тествоых данных. на следующем шаге разделю на тренировачные и валидационные по полам
# random_state=42 - это для того чтобы результаты были воспроизводимыми
# X_train, X_test, y_train, y_test  надо запомнить этот порядок
X_train, X_test, y_train, y_test =  train_test_split(feutures, labels, test_size=0.4, random_state=42) 


# разделяю тестовые данные на валидационные и тестовые. тепер уже вместо feutures и labels использую X_test и y_test
X_test, X_val, y_test, y_val =  train_test_split(X_test, y_test, test_size=0.5, random_state=42) 


# check data allocations
print(len(labels), len(y_train), len(y_val), len(y_test))
```
