# bar plot

```python
exam_means  = pd.DataFrame({'exam':['exam1', 'exam2', 'exam3', 'exam4'], 'mean':[50, 60, 70, 80]})
exam_means
```

```python
plt.bar(exam_means['exam'], exam_means['mean'])
plt.ylabel('Mean Score')
plt.xlabel('Exams')
plt.title('Mean Score of 4 Exams')
plt.show()
```

```python
# pie chart
categ = ['Homework', 'Labs', 'Quizzes', 'Final Exam', 'Midterm']

weights = [20, 10, 10, 30, 30] # in %

breakdown = pd.DataFrame({'categories':categ, 'weights':weights})

plt.figure(figsize=(7,7))
plt.pie(breakdown['weights'], labels=breakdown['categories'], autopct='%1.1f%%')
plt.title('Grade Breakdown')
plt.show()
```

```python
# histogram
exam1_grades = pd.Series(np.random.normal(50, 15, 400))

plt.hist(exam1_grades, bins=20, color='skyblue')
plt.ylabel('Number of Students')
plt.xlabel('Grades')
plt.title('Exam 1 Grades')
plt.show()
```

```python
# subplot

exam1_grades = pd.Series(np.random.normal(50, 15, 400))
exam2_grades = pd.Series(np.random.normal(60, 10, 400))

# строю граффики
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
axes[0].hist(exam1_grades, bins=20, color='skyblue')
axes[0].set_title('Exam 1 Grades')

axes[1].hist(exam2_grades, bins=20, color='skyblue')
axes[1].set_title('Exam 2 Grades')

plt.show()

```

```python
# kernel density plot график плотности ядра
# оценки 400 студентов
scores = pd.Series(np.random.normal(50, 10, 400))

mean = [60, 70] # среднее значение

covariance_matrxi = [(1, .5), (5., 1)] # матрица ковариации это когда одна переменная зависит от другой. изменения значения относительно друг друга

data = np.random.multivariate_normal(mean, covariance_matrxi, size=400) # генерирую данные

midterm_scores = pd.DataFrame(data, columns=['Midterm1', 'Midterm2'])

midterm_scores.head()
```

```python
# график плотности ядра для двух переменных
sns.kdeplot(x=midterm_scores['Midterm1'], y=midterm_scores['Midterm2'], shade=False)
plt.show()
```
