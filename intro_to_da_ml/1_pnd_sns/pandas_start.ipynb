```python
import pandas as pd
```

```python
file = '../data/StudentsPerformance.csv'
```

```python
student_perf = pd.read_csv(file)
student_perf.head()
```

# группировка

```python
student_perf.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
```

```python
student_perf.groupby('gender').aggregate({'math score': 'mean', 'reading score': 'mean', 'writing score': 'mean'})
```

```python
student_perf.groupby('gender', as_index=False)\
    .aggregate({'math score': 'mean', 'reading score': 'mean', 'writing score': 'mean'})\
    .rename(columns={'math score': 'mean_math_score', 'reading score': 'mean_reading_score', 'writing score': 'mean_writing_score'})
```

```python
# если убрать as_index=False, то получим мультииндекс. не очень удобно
student_perf.groupby(['gender', 'race/ethnicity'], as_index=False)\
    .aggregate({'math score': 'mean', 'reading score': 'mean', 'writing score': 'mean'})\
    .rename(columns={'math score': 'mean_math_score', 'reading score': 'mean_reading_score', 'writing score': 'mean_writing_score'})
```

```python
# если убрать as_index=False, то получим мультииндекс. не очень удобно
mean_scores = student_perf.groupby(['gender', 'race/ethnicity'])\
    .aggregate({'math score': 'mean', 'reading score': 'mean', 'writing score': 'mean'})\
    .rename(columns={'math score': 'mean_math_score', 'reading score': 'mean_reading_score', 'writing score': 'mean_writing_score'})
```

```python
concentrations = pd.read_csv('../data/tarea.csv')
concentrations.head()
```

```python
concentrations.groupby(['genus'], as_index = False)\
    .aggregate(
    mean_suc = ('sucrose', 'mean'),
    mean_alanin = ('alanin', 'mean'),
    mean_citrate = ('citrate', 'mean'),
    mean_glucose = ('glucose', 'mean'),
    mean_oleic_acid = ('oleic_acid', 'mean')
)

```

```python
concentrations.groupby(['genus']).mean(numeric_only=True)
```

```python
fucus_data = concentrations[concentrations.genus == 'Fucus']
fucus_data
```

```python
min_alanin = round(fucus_data['alanin'].min(), 2)
mean_alanin = round(fucus_data['alanin'].mean(), 2)
max_alanin = round(fucus_data['alanin'].max(), 2)

result = f"{min_alanin} {mean_alanin} {max_alanin}"
result
```

```python
concentrations.groupby('genus').aggregate({'alanin': ['min', 'mean', 'max']}).loc[['Fucus']]
```
