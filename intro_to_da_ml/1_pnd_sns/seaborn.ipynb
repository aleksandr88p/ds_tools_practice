```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_ = '../data/StudentsPerformance.csv'
df = pd.read_csv(file_)
df.head()
```

```python
df['math score'].mean()
```

```python
# самый простой способ построить график и посмотерть на распределение например math score

df['math score'].hist()
```

```python
# scatter plot нужен для того чтобы посмотреть на взаимосвязь между двумя переменными

df.plot.scatter(x='math score', y='reading score')
```

```python
# если надо как то стилизовать график, то можно использовать seaborn
# hue группирует данные по какому то признаку, например по полу
# fit_reg=True строит линию регрессии, чтобы посмотреть на график
sns.lmplot(x='math score', y='reading score', hue='gender', data=df, fit_reg=True)
```

```python
# если sns граффик добавить в переменную, то можно обращаться к его методам и менять например названия осей
ax = sns.lmplot(x='math score', y='reading score', hue='gender', data=df, fit_reg=True)
ax.set_xlabels('Math score of student')
ax.set_ylabels('Reading score of student')
ax.set_titles('Math score vs Reading score')
ax.legend.set_title('Gender here')
```
