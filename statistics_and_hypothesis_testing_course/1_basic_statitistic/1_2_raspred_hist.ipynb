```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import random as random
```

```python
np.random.seed(5) #создаем идемпотентный рандом
class1 = np.random.randint(30, 100, 30)  #выборка с целыми числами от 1 до 100
class2 = np.random.randint(30, 100, 30)  #выборка с целыми числами от 1 до 100
```

```python
plt.hist(class1, bins=5) #cоздаем гистограмму
plt.grid(10) #рисуем сетку на график
# plt.xticks(np.arange(min(data), max(data)+1,1)) #подписи по оси X
# plt.yticks([1,2,3,4]) #подписи по оси Y
plt.show() #показываем гистограмму
```

```python
plt.hist(class2, bins=10) #cоздаем гистограмму
plt.show() #показываем гистограмму
```

Любым способом (в идеале с помощью pandas) посчитайте для каждой уникальной оценки (для каждого балла) из списка class1 кол-во повторений. Какой балл повторяется чаще всего.

```python
class1
```

```python
# делаю датафрейм из одного класса
df_class1 = pd.DataFrame(class1, columns=['class1'])
df_class1.groupby('class1').size().idxmax() #группируем по классам, считаем количество, находим максимальное
```

```python
df_class1['class1'].value_counts().idxmax() #находим максимальное значение класса в датафрейме
```

```python
values, counts = np.unique(class1, return_counts=True) # метод unique возвращает уникальные значения и их количество
values[counts.argmax()] # потом обращаемся к индексу максимального значения
```
