```python
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats as sts
```

Для дискретной случайной величины мы можем посчитать количество значений для каждого события и найти вероятность каждого события. Можем представить случайную величину в виде таблицы. Непрерывную случайную величины нельзя представить в виде таблицы, так как невозможно перечислить все значения и расчитать соответствующие им вероятности, потому что значений бесконечно много.

Непрерывную величину обычно задают с помощью двух функций: **функция плотности вероятностей** и **функция распределеления вероятностей**

### 1.1 Функция плотности вероятностей

на aнглийском **Probability density function**. Сокращенно **PDF**

Когда мы говорили про дискретные случайные величины, мы расчитывали **закон распределения** и строили **многоугольник распределения**. У непрерывных случайных величин есть аналог - **функция плотности**

Давайте сразу построим график функции плотности вероятности для некоторого распределения, которое называют **нормальным**. Нормальное распределение - это одно из важнейших распределений в статистике, его мы разберем позже.

Есть несколько известных дискретных и непрерывных распределений, которые имеют свои уникальные характеристики и применение в статистике. Некоторые из них мы рассмотрим.

```python
#задаем непрерывное нормальное распределение с математическим ожиданием 0 и стандартным отклонением 1
mu = 0
sigma = 1
norm_rv = sts.norm(loc=mu, scale=sigma) #задаем параметры распределения
X = norm_rv.rvs(size=1000) #задаем количество значений
```

```python
X[:10]
```

```python
#строим гистограмму для визуализации распределения
plt.hist(X, bins=30) #указываем количество бинов (диапазонов)
plt.show()
```

```python
#с помощью параметра density=True можно поменять по оси Y кол-во значений на вероятность
plt.hist(X, bins=30, density=True)
plt.show()
```

Теперь построим саму функцию плотности вероятности

```python
plt.hist(X, bins=30, density=True) #строим гистограмму

x = np.linspace(-4,4,1000) #cоздает массив из 1000 равномерно-распределенных точек. 
#Кстати про равномерное распределение мы тоже поговорим позднее, на следующем уроке.
pdf = norm_rv.pdf(x)#вычисляем функцию плотности вероятности нормального распределения для каждого значения x

plt.plot(x, pdf,color='red')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

Площадь под функцией плотности вероятности равна единице

Мы не можем расчитать вероятность отдельно взятого значения, потому что их бесконечно много.</br>
Но мы можем расчитать вероятность попадания значения в определенный диапозон значения.

### 1.2 Функция распределения вероятностей

На английском **Cumulative distribution function** или **CDF**

Зная как задается функция плотности распределения вероятностей **PDF**, мы можем построить функцию распределения вероятностей **CDF**. И наборот.</br>
С функцией распределения вероятностей мы уже встречались, когда говорили о дисткретных случайных величинах.

Формула такая же как и для дискретных случайных величин: $F(x)=P(X \leq x)$

, где: $F(x)$ – это вероятность $P$ того, что случайная величина $X$ примет значение, меньшее или равное, чем переменная (значение) $x$.

```python
x = np.linspace(-4,4,1000)
cdf = norm_rv.cdf(x) #для каждого x расчитываем CDF
plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.show()
```

Напонимаю, что функция распределения вероятностей неубывающая. Но в отличие от дискретной случайной величины, функция гладкая, а не разрывная-ступенчатая

### 1.3 графики PDF и CDF

Нарисуем графики PDF и CDF рядом и посмотрим связь между ними

```python
fig, axs = plt.subplots(nrows = 2, ncols=1) #указываем, что графика 2, в одном столбце

#PDF
x = np.linspace(-4,4,1000)
pdf = norm_rv.pdf(x)
axs[0].plot(x, pdf)
axs[0].grid(True)


#CDF
cdf = norm_rv.cdf(x)
axs[1].plot(x, cdf)
axs[1].grid(True)

plt.show()
```

Как я упомянул выше, зная PDF, можно построить CDF. Для этого нужно проинтегрировать функцию PDF. </br>
Зная функцию CDF, можно посторить PDF, продифференцировав функцию CDF.
