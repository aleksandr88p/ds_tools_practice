```python
from scipy import stats as sts
import numpy as np
from matplotlib import pyplot as plt
```

Равномерное распределение бывает как дискретным, так и непрерывным.

### 1. Дискретное равномерное распределение

Расмотрим, как задавать дискретное равномерное распределение.

```python
#создаем распределение целых чисел от 0 до 9
X = [np.random.randint(0,10) for i in range(1000)]
```

```python
X[0:10]
```

```python
len(X)
```

```python
#построим гистограмму
plt.hist(X)
plt.show()
```

```python
#построим функцию плотности вероятности
plt.hist(X, density = True)
plt.plot()
```

**Равномерное дискретное распределение** — это распределение, в котором каждое возможное значение случайной величины X имеет одинаковую вероятность и лежит в пределах от a до b, где a - минимальное значение, b - максимальное значение

**X ∼ U(a, b)**

для равномерного распределения также будет справедлива формула вероятности:

$P(A) = \frac m n$



где:
$(A)$ - это некторое событие;
$m$ - это число исходов благоприятствующих событию А;
$n$ - это число всех исходов.

Мы можем посчитать вероятность того, что случайная величина примет значение 0, затем 1, затем 2 и так далее ... </br>
получится что вероятности каждого уникального значения будут равны $P(X=0) = P(X=1) = ... = P(X=10)$

Можно переписать формулу вероятности для каждого уникального значения: 

$P (X = x) =\frac1 n$


### 2. Функция распределения вероятности для дискретного равномерного распределения

Функция распределения вероятности для всех уникальных значений будет иметь вид: $F (x) = \frac{ x_i−a+1} n$ <br>
где:</br>
a - это минимальное значение в выборке</br>
n - это кол-во элементов в выборке

```python
a = 0
```

```python
n = 10
```

```python
np.unique(X)
```

```python
F_x = []
for x in np.unique(X):
    F_x.append((x-a+1)/n)
```

```python
plt.step(np.unique(X), F_x, where='post')
plt.show()
```

### 3. Матожидание и дисперсия дискретного равномерного распределения

```python
np.mean(X) #среднее
```

```python
#дисперсия
np.var(X)
```

```python
np.std(X) #стандартное отклонение
```

Свойства теоретического равномерного распределения:</br>

математическое ожидание: $M(X) = \frac {𝑎+𝑏} 2$

Дисперсия: $ Var(X) = \frac{(𝑏−𝑎)^2} {12}$

### 4. Непрерывное равномерное распределение

Рассмотрим как задавать непрерывное равномерное распределение

Построим распределение **X ∼ U(0, 1)** двумя способами

1 способ - с помощью библиотеки numpy

```python
X = [np.random.random() for i in range(1000)]
```

```python
plt.hist(X)
plt.show()
```

2 способ - c помощью библитеки scipy

```python
X = sts.uniform.rvs(size = 1000)
```

```python
X[:10]
```

```python
plt.hist(X)
plt.show()
```

Построим распределение **X ∼ U(a, b)**, зададим a и b

Для равномерного распределения необходимо задать параметры:</br>
Параметр $loc = a$ определяет начало отрезка, на котором случайная величина равномерно распределена.</br>
Параметр  $scale = b-a$ определяет длину отрезка, на котором задана случайная величина.</br>
Значение плотности распределения на данном отрезке убывает с ростом данного параметра, то есть с ростом длины этого отрезка. Чем меньше длина отрезка, тем больше значение плотности вероятности на отрезке.

```python
X = sts.uniform.rvs(loc = 0, scale = 20,size = 1000) #a - это loc, b - это loc+scale
```

```python
X[0:10]
```

```python
plt.hist(X)
plt.show()
```

```python
plt.hist(X, density = True)
plt.show()
```

**Построим функцию плотности вероятностей**

```python
# Построение гистограммы для визуализации распределения
count, bins, ignored = plt.hist(X, 10, density=True)

# Построение графика функции плотности вероятности
plt.plot(bins, np.full_like(bins, 1/20), linewidth=2, color='r')

# Настройка отображения
plt.title('График функции плотности вероятности для равномерного распределения')
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')
plt.grid(True)
plt.show()
```

**Задача на PDF**: Найти вероятность плотность вероятности значений от 0 до 5

```python
sts.uniform.cdf(5,0,20)
```

**Построим функию распределения вероятностей**

```python
# Построение графика функции распределения вероятности
plt.hist(X, bins=10, density=True, cumulative=True)

# Настройка отображения
plt.title('График функции распределения вероятности для равномерного распределения')
plt.plot(np.sort(X), np.linspace(0, 1, X.size), color='r')
plt.xlabel('Значение')
plt.ylabel('Вероятность')
plt.grid(True)
plt.show()
```

**Задача на CDF**: найти значение которое соответсвует 0.4 квантилю

```python
#с помощью  ppf(percent point function) можем найти значение которое соответсвует 0.4 квантилю
sts.uniform.ppf(0.4,0,20)
```
