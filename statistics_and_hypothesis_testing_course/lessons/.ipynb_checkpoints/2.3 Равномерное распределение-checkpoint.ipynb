```python
from scipy import stats as sts
import numpy as np
from matplotlib import pyplot as plt
```

–†–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –±—ã–≤–∞–µ—Ç –∫–∞–∫ –¥–∏—Å–∫—Ä–µ—Ç–Ω—ã–º, —Ç–∞–∫ –∏ –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã–º.

### 1. –î–∏—Å–∫—Ä–µ—Ç–Ω–æ–µ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ

–†–∞—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫ –∑–∞–¥–∞–≤–∞—Ç—å –¥–∏—Å–∫—Ä–µ—Ç–Ω–æ–µ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ.

```python
#—Å–æ–∑–¥–∞–µ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª –æ—Ç 0 –¥–æ 9
X = [np.random.randint(0,10) for i in range(1000)]
```

```python
X[0:10]
```

```python
len(X)
```

```python
#–ø–æ—Å—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É
plt.hist(X)
plt.show()
```

```python
#–ø–æ—Å—Ç—Ä–æ–∏–º —Ñ—É–Ω–∫—Ü–∏—é –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏
plt.hist(X, density = True)
plt.plot()
```

**–†–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–µ –¥–∏—Å–∫—Ä–µ—Ç–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ** ‚Äî —ç—Ç–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ, –≤ –∫–æ—Ç–æ—Ä–æ–º –∫–∞–∂–¥–æ–µ –≤–æ–∑–º–æ–∂–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã X –∏–º–µ–µ—Ç –æ–¥–∏–Ω–∞–∫–æ–≤—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –∏ –ª–µ–∂–∏—Ç –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö –æ—Ç a –¥–æ b, –≥–¥–µ a - –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ, b - –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ

**X ‚àº U(a, b)**

–¥–ª—è —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ç–∞–∫–∂–µ –±—É–¥–µ—Ç —Å–ø—Ä–∞–≤–µ–¥–ª–∏–≤–∞ —Ñ–æ—Ä–º—É–ª–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏:

$P(A) = \frac m n$



–≥–¥–µ:
$(A)$ - —ç—Ç–æ –Ω–µ–∫—Ç–æ—Ä–æ–µ —Å–æ–±—ã—Ç–∏–µ;
$m$ - —ç—Ç–æ —á–∏—Å–ª–æ –∏—Å—Ö–æ–¥–æ–≤ –±–ª–∞–≥–æ–ø—Ä–∏—è—Ç—Å—Ç–≤—É—é—â–∏—Ö —Å–æ–±—ã—Ç–∏—é –ê;
$n$ - —ç—Ç–æ —á–∏—Å–ª–æ –≤—Å–µ—Ö –∏—Å—Ö–æ–¥–æ–≤.

–ú—ã –º–æ–∂–µ–º –ø–æ—Å—á–∏—Ç–∞—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞ –ø—Ä–∏–º–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ 0, –∑–∞—Ç–µ–º 1, –∑–∞—Ç–µ–º 2 –∏ —Ç–∞–∫ –¥–∞–ª–µ–µ ... </br>
–ø–æ–ª—É—á–∏—Ç—Å—è —á—Ç–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–∞–∂–¥–æ–≥–æ —É–Ω–∏–∫–∞–ª—å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –±—É–¥—É—Ç —Ä–∞–≤–Ω—ã $P(X=0) = P(X=1) = ... = P(X=10)$

–ú–æ–∂–Ω–æ –ø–µ—Ä–µ–ø–∏—Å–∞—Ç—å —Ñ–æ—Ä–º—É–ª—É –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —É–Ω–∏–∫–∞–ª—å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è: 

$P (X = x) =\frac1 n$


### 2. –§—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è –¥–∏—Å–∫—Ä–µ—Ç–Ω–æ–≥–æ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è

–§—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è –≤—Å–µ—Ö —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –±—É–¥–µ—Ç –∏–º–µ—Ç—å –≤–∏–¥: $F (x) = \frac{ x_i‚àía+1} n$ <br>
–≥–¥–µ:</br>
a - —ç—Ç–æ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤ –≤—ã–±–æ—Ä–∫–µ</br>
n - —ç—Ç–æ –∫–æ–ª-–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –≤—ã–±–æ—Ä–∫–µ

```python
a = 0
```

```python
n = 10
```

```python
np.unique(X)
```

```python
F_x = []
for x in np.unique(X):
    F_x.append((x-a+1)/n)
```

```python
plt.step(np.unique(X), F_x, where='post')
plt.show()
```

### 3. –ú–∞—Ç–æ–∂–∏–¥–∞–Ω–∏–µ –∏ –¥–∏—Å–ø–µ—Ä—Å–∏—è –¥–∏—Å–∫—Ä–µ—Ç–Ω–æ–≥–æ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è

```python
np.mean(X) #—Å—Ä–µ–¥–Ω–µ–µ
```

```python
#–¥–∏—Å–ø–µ—Ä—Å–∏—è
np.var(X)
```

```python
np.std(X) #—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
```

–°–≤–æ–π—Å—Ç–≤–∞ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–≥–æ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è:</br>

–º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ: $M(X) = \frac {ùëé+ùëè} 2$

–î–∏—Å–ø–µ—Ä—Å–∏—è: $ Var(X) = \frac{(ùëè‚àíùëé)^2} {12}$

### 4. –ù–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ–µ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –∫–∞–∫ –∑–∞–¥–∞–≤–∞—Ç—å –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ–µ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ

–ü–æ—Å—Ç—Ä–æ–∏–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ **X ‚àº U(0, 1)** –¥–≤—É–º—è —Å–ø–æ—Å–æ–±–∞–º–∏

1 —Å–ø–æ—Å–æ–± - —Å –ø–æ–º–æ—â—å—é –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ numpy

```python
X = [np.random.random() for i in range(1000)]
```

```python
plt.hist(X)
plt.show()
```

2 —Å–ø–æ—Å–æ–± - c –ø–æ–º–æ—â—å—é –±–∏–±–ª–∏—Ç–µ–∫–∏ scipy

```python
X = sts.uniform.rvs(size = 1000)
```

```python
X[:10]
```

```python
plt.hist(X)
plt.show()
```

–ü–æ—Å—Ç—Ä–æ–∏–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ **X ‚àº U(a, b)**, –∑–∞–¥–∞–¥–∏–º a –∏ b

–î–ª—è —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∑–∞–¥–∞—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:</br>
–ü–∞—Ä–∞–º–µ—Ç—Ä $loc = a$ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –Ω–∞—á–∞–ª–æ –æ—Ç—Ä–µ–∑–∫–∞, –Ω–∞ –∫–æ—Ç–æ—Ä–æ–º —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∞.</br>
–ü–∞—Ä–∞–º–µ—Ç—Ä  $scale = b-a$ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –¥–ª–∏–Ω—É –æ—Ç—Ä–µ–∑–∫–∞, –Ω–∞ –∫–æ—Ç–æ—Ä–æ–º –∑–∞–¥–∞–Ω–∞ —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞.</br>
–ó–Ω–∞—á–µ–Ω–∏–µ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–∞ –¥–∞–Ω–Ω–æ–º –æ—Ç—Ä–µ–∑–∫–µ —É–±—ã–≤–∞–µ—Ç —Å —Ä–æ—Å—Ç–æ–º –¥–∞–Ω–Ω–æ–≥–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞, —Ç–æ –µ—Å—Ç—å —Å —Ä–æ—Å—Ç–æ–º –¥–ª–∏–Ω—ã —ç—Ç–æ–≥–æ –æ—Ç—Ä–µ–∑–∫–∞. –ß–µ–º –º–µ–Ω—å—à–µ –¥–ª–∏–Ω–∞ –æ—Ç—Ä–µ–∑–∫–∞, —Ç–µ–º –±–æ–ª—å—à–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ.

```python
X = sts.uniform.rvs(loc = 0, scale = 20,size = 1000) #a - —ç—Ç–æ loc, b - —ç—Ç–æ loc+scale
```

```python
X[0:10]
```

```python
plt.hist(X)
plt.show()
```

```python
plt.hist(X, density = True)
plt.show()
```

**–ü–æ—Å—Ç—Ä–æ–∏–º —Ñ—É–Ω–∫—Ü–∏—é –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π**

```python
# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
count, bins, ignored = plt.hist(X, 10, density=True)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏
plt.plot(bins, np.full_like(bins, 1/20), linewidth=2, color='r')

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è
plt.title('–ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è')
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ')
plt.ylabel('–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏')
plt.grid(True)
plt.show()
```

**–ó–∞–¥–∞—á–∞ –Ω–∞ PDF**: –ù–∞–π—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–ª–æ—Ç–Ω–æ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∑–Ω–∞—á–µ–Ω–∏–π –æ—Ç 0 –¥–æ 5

```python
sts.uniform.cdf(5,0,20)
```

**–ü–æ—Å—Ç—Ä–æ–∏–º —Ñ—É–Ω–∫–∏—é —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π**

```python
# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏
plt.hist(X, bins=10, density=True, cumulative=True)

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è
plt.title('–ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è')
plt.plot(np.sort(X), np.linspace(0, 1, X.size), color='r')
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ')
plt.ylabel('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å')
plt.grid(True)
plt.show()
```

**–ó–∞–¥–∞—á–∞ –Ω–∞ CDF**: –Ω–∞–π—Ç–∏ –∑–Ω–∞—á–µ–Ω–∏–µ –∫–æ—Ç–æ—Ä–æ–µ —Å–æ–æ—Ç–≤–µ—Ç—Å–≤—É–µ—Ç 0.4 –∫–≤–∞–Ω—Ç–∏–ª—é

```python
#—Å –ø–æ–º–æ—â—å—é  ppf(percent point function) –º–æ–∂–µ–º –Ω–∞–π—Ç–∏ –∑–Ω–∞—á–µ–Ω–∏–µ –∫–æ—Ç–æ—Ä–æ–µ —Å–æ–æ—Ç–≤–µ—Ç—Å–≤—É–µ—Ç 0.4 –∫–≤–∞–Ω—Ç–∏–ª—é
sts.uniform.ppf(0.4,0,20)
```
