### 1. Определения

**Генеральная совокупность** - это полный набор всех элементов, которые нас интересуют в контексте исследования. Например, если мы интересуемся ростом всех людей в стране, то генеральная совокупность будет включать в себя рост каждого человека в этой стране.

**Выборка (sample)** - это подмножество элементов из генеральной совокупности, которое анализируются для получения знаний о генеральной совокупности и принятия выводов о ней на основе статистических методов.

Выборка должна быть репрезентативной для генеральной совокупности, чтобы статистические выводы, сделанные на ее основе, могли быть обобщены на всю генеральную совокупность. Корректное формирование выборки и адекватное представление генеральной совокупности важно для правильного проведения статистических исследований.

### 2. Способы формирования выборок из генеральной совокупности

1. Простая случайная выборка: Каждый элемент имеет одинаковую вероятность быть выбранным, простым случайным образом.

2. Стратифицированная выборка: Генеральная совокупность делится на более мелкие подгруппы (страты), и из каждой страты случайным образом выбирается образец. Этот метод может быть более представительным, чем простая случайная выборка, особенно если есть большие различия между стратами.

3. Систематическая выборка: Элементы выбираются через постоянные интервалы из генеральной совокупности. Например, каждый 10-й элемент может быть выбран для включения в выборку.

4. Кластерная выборка: Генеральная совокупность делится на группы или кластеры, и случайно выбираются целые кластеры для включения в выборку.

Выбор метода выборки зависит от целей исследования, доступности информации и стоимости сбора данных.

Статистическая гипотеза  − любое предположение о распределении и свойствах случайной величины.

### 3. Точечные оценки

Точечная оценка в статистике представляет собой оценку неизвестного параметра ГС (например, среднего значения или доли) на основе полученных выборочных данных. Это может быть конкретное значение, которое предполагается как лучшая оценка для параметра.

Например, если мы хотим оценить среднюю продолжительность сна в Токио, мы можем взять выборку жителей Токио и рассчитать среднюю продолжительность сна этой выборки. Это значение средней продолжнительности сна в выборке будет точечной оценкой средней продолжнительности сна в Токио.

Точечная оценка не дает информации о достоверности оценки и не учитывает возможные ошибки, связанные с выборкой. Для более полной картины используется интервальная оценка, которая учитывает неопределенность и позволяет оценить доверительный интервал для параметра с учетом прогнозируемых ошибок.

Точечные оценки должны обладать некоторыми свойствами: 
1. **Несмещенность:** Точечная оценка называется несмещенной, если ожидаемое значение оценки равно истинному значению оцениваемого параметра. Другими словами, несмещенная оценка не искажает истинное значение параметра.

2. **Состоятельность:** Точечная оценка называется состоятельной, если она сходится к оцениваемому параметру по мере увеличения размера выборки. Это свойство гарантирует, что с увеличением объема выборки точечная оценка приближается к истинному значению параметра.

3. **Эффективность**: Точечная оценка называется эффективной, если она имеет наименьшую дисперсию среди всех несмещенных оценок. Это означает, что она обладает наилучшей точностью относительно других несмещенных оценок.

4. **Доверительный интервал**: Желательно, чтобы точечная оценка была сопровождена доверительным интервалом, который предоставляет информацию о диапазоне значений, в котором, с заданной вероятностью, содержится истинное значение параметра.

5. **Предельная нормальность**: В больших выборках точечные оценки хорошо апроксимируются нормальным распределением вокруг истинного значения параметра.

**Несмещенность**

Точечная оценка называется несмещенной, если ожидаемое значение оценки равно истинному значению оцениваемого параметра. Другими словами, несмещенная оценка не искажает истинное значение параметра.

Среднее по выборке является хорошей оценкой среднего генеральной совокупности

```python
from scipy import stats as sts
import numpy as np
from matplotlib import pyplot as plt
import statistics as st
```

```python
mu = 10
sigma = 15
norm_rv = sts.norm(loc=mu, scale=sigma) #задаем параметры распределения
X = norm_rv.rvs(size=100000) #создаем генеральную совокупность из 100 000 элементов
```

```python
# Визуализация распределения ГС
plt.hist(X, bins=30)

# Отображение графика
plt.title('Гистограмма нормального распределения')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.show()
```

$ \bar{x} = \frac 1 {n}{\sum\limits_{i=1}^{n} x_i} $

где: <br/> 
$ \bar{x} $- среднее;<br/> 
$ n $- количесво наблюдений; <br/>  
$ {\sum\limits_{i=1}^{n}} $ -сумма всех наблюдений от 1 до n 

```python
#среднее генеральной совокупности близко к 10
np.mean(X)
```

```python
#возьмем тысячу значений из ГС и посчитаем среднее
```

```python
#выборка
sample = np.random.choice(X, size=1000, replace=False)
```

```python
np.mean(sample) #в среднем оценка среднего по выборке = среднему генеральной совокупности
```

```python
#проведем эксперемент 20 раз
list_means = []
for i in range(20):
    sample = np.random.choice(X, size=1000, replace=False)
    list_means.append(np.round(np.mean(sample),2))
```

```python
plt.figure(figsize=(20,6))
plt.plot(list_means,'ro',linewidth=1.0, label="Graph2")
plt.xticks(np.arange(len(list_means)), np.arange(1, len(list_means)+1))
plt.plot()
```

**Пример смещенной оценки**

Вспомним формулу дисперсии и формулу дисперсии для выборки.

$ Var(X) = \frac 1{n}{\sum\limits_{i=1}^{n}(x_i}-\bar{x})^2  $

$ Var(X) = \frac 1{n-1}{\sum\limits_{i=1}^{n}(x_i}-\bar{x})^2  $

где:<br/>
$X$ - это группа наблюдений, далее будем называть ее выборкой;<br/> 
$Var(X)$ - это дисперсия группы наблюдений;<br/> 
$ {\sum\limits_{i=1}^{n}} $ - сумма всех наблюдений от 1 до n;<br/> 
$(x_i-\bar{x})^2 $ - квадрат разности между наблюдением и средним;

Вспомним формулу стандартного отклонения

$ \sqrt{Var(X)} $

Зачем в формуле для выборки отнимать единицу от n? Ответ - чтобы убрать смещение.

```python
list_std_biased = [] #смещенные оценки
list_std_unbiased = []  #несмещенные оценки
for i in range(20):
    sample = np.random.choice(X, size=10, replace=False)
    list_std_biased.append(np.std(sample, ddof=0))
    list_std_unbiased.append(np.std(sample, ddof=1)) #корректировка смещения
```

```python
plt.figure(figsize=(20,6))
plt.plot(list_std_biased,'ro',linewidth=1.0, label="Graph2")
plt.xticks(np.arange(len(list_std_biased)), np.arange(1, len(list_std_biased)+1))
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.plot(list_std_unbiased,'ro',linewidth=1.0, label="Graph2")
plt.xticks(np.arange(len(list_std_unbiased)), np.arange(1, len(list_std_unbiased)+1))
plt.show()
```

```python
#среднее стандартных смещенных отклонений
np.mean(list_std_biased)
```

```python
#среднее стандартных несмещенных оценок
np.mean(list_std_unbiased)
```

```python
#стандартное отклонение ГС
np.std(X)
```

Если мы будем увеличивать кол-во элементов в выборке, то смещение будет стремиться к нулю. Оценка обладающая такими свойствами называется асимптотически несмещенной. И для статистика это хорошая оценка.

**Предельная нормальность и доверительный интервал**

Желательно, чтобы точечная оценка была сопровождена доверительным интервалом, который предоставляет информацию о диапазоне значений, в котором, с заданной вероятностью, содержится истинное значение параметра.

```python
sample = np.random.choice(X, size=1000, replace=False)
```

```python
np.mean(sample)
```

```python
#проведем эксперемент 1000 раз
list_means = []
for i in range(1000):
    sample = np.random.choice(X, size=1000, replace=False)
    list_means.append(np.round(np.mean(sample),2))
```

```python
# Визуализация гистограммы
plt.hist(list_means, bins=30)

# Отображение графика
plt.title('Гистограмма выборочного среднего')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.show()
```

**Центральная предельная теорема (ЦПТ)** - если многократно извлекать выборки из ГС, то выборочное среднее будет распределено нормально с математическим ожиданием $\mu$ и дисперсией  $\sigma^2$

```python
mean_of_means = np.mean(list_means)
print(mean_of_means)
```

```python
np.std(list_means)
```

```python
std_of_means = np.std(list_means)/len(list_means)**0.5
print(std_of_means)
```

```python
#95%-й доверительный интервал
print(mean_of_means - 1.96 * std_of_means, mean_of_means + 1.96 * std_of_means)
```

**Состоятельность**

```python
#размеры выборорок
sizes = [10, 50, 100, 200, 1000]
```

```python
mean_of_means = []
std_of_means = []
for size in sizes:
    list_means = []
    for i in range(1000):
        sample = np.random.choice(X, size=size, replace=False)
        list_means.append(np.mean(sample))
    mean_of_means.append(np.mean(list_means))
    std_of_means.append(np.std(list_means))
```

```python
lower = []
upper = []
for mean, std in zip(mean_of_means, std_of_means):
    lower.append(mean - 1.96 * std)
    upper.append(mean + 1.96 * std)
```

```python
#поправить код, перепроверить
plt.plot(sizes, mean_of_means, label='mean means')
plt.fill_between(sizes, lower, upper, alpha=0.3, label='95% доверительный интервал')
plt.xscale('log')
plt.title('Сходимость среднего средних')
plt.xlabel('size')
plt.legend()
plt.grid()
plt.show()
```

**Закон больших чисел (ЗБЧ)** - основная идея заключается в том, что среднее значение независимых и одинаково распределенных случайных величин сходится к математическому ожиданию этого распределения по мере увеличения размера выборки.
