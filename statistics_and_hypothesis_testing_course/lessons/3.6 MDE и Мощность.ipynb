```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats as sts
```

### 1. Ошибка второго рода

Вы уже знакомы с понятием ошибка первого рода. Ошибка первого рода - это вероятность отклонить нулевую гипотезу, когда она верна. Иными словами, ошибка первого рода - это вероятность найти различие, там где различия нет.</br>
В случае двухвыборочного теста ошибка первого рода - это когда мы ошибочно говорим, что выборки отличаются, хотя на самом деле они не отличаются.

При проведении эксперемента можно совершить 2 ошибки: ошибку первого рода и ошибку второго рода.

Ошибка второго рода - это вероятность **не отклонить** нулевую гипотезу в случае, когда **нулевая гипотеза неверна**.</br>
Ошибка второго рода - это вероятность **не найти различия**, если **различие действительно есть**</br>
Ошибка второго рода - это вероятность сказать, что выборки **не отличаются, хотя они отличаются**

Приведу примеры экспериментов, когда важно минимизировать ошибку первого рода, а когда ошибку второго рода.

**Пример 1 (важна ошибка первого рода)**:</br>
Ученые некоторой компании создали лекарство. Ученым-статистикам необходимо доказать действенность этого лекарства. Для этого они набрали группу больных, разделили их на тестовую и контрольную группу. Тестовая группа приняла лекарство, а контрольная - нет. 

H0 - разницы между группами нет. Лекарство не работает.
H1 - разница между группами есть. Лекарство работает.

Ученым важно, чтобы лекарство действительно работало почти безоговорочно. В этом случае важна ошибка первого рода, она должна быть как можно ниже. Чем ниже ошибка первого рода, ниже вероятность ошибочно найти различия.

**Пример 2 (важна ошибка второго рода):**</br>


Например, при тестировании на рак, ошибочное необнаружение рака может иметь более серьезные последствия, чем ложная тревога об его наличии.</br>
Допустимую вероятность ошибки первого рода обозначают латинской буквой альфа $\alpha$</br>
Допустимую вероятность ошибки второго рода обозначают латинской буквой бета $\beta$. </br>
Бета также как и альфа задается перед запуском эксперимента</br>
В бизнес индустрии чаще выбирают $\alpha$ = 0.05, $\beta$ = 0.2

### 2. Зависимость ошибки первого рода и второго рода

Если многократно брать пары выборок из двух ГС, считать разность выборочных средних для этих выборок, то мы получим нормальное распределение с параметрами:</br>
$\mu$ = $\mu_{test}$ - $\mu_{control}$ и $Var = \frac{Var_{test}}{n_{test}}+\frac{Var_{control}}{n_{control}}$

При этом если будет верна H0, то среднее распределение разности выборочных средних будет равно 0.

Изобразим распределения разности выборочных средних при верности нулевой гипотезы и при верности альтернативной гипотезы на графике.

```python
plt.figure(figsize=(12, 6), dpi=80)

norm_rv = sts.norm(loc=0, scale=1)
norm_rv_2 = sts.norm(loc=3, scale=1)

x = np.linspace(-4,8,1000) #cоздает массив из 1000 равномерно-распределенных точек. 
pdf = norm_rv.pdf(x)#вычисляем функцию плотности вероятности нормального распределения для каждого значения x
pdf_2 = norm_rv_2.pdf(x)

plt.plot(x, pdf, color ='C0')
plt.plot(x, pdf_2, color ='C0')
plt.fill_between(x, pdf, where=(x >= -4) & (x <= -1.96), color ='C0')

plt.fill_between(x, pdf_2, where=(x >= -4) & (x <= 2), color ='C2', alpha=0.5)
plt.fill_between(x, pdf_2, where=(x >= 2) & (x <= 6), color ='C1',alpha=0.5)

plt.fill_between(x, pdf, where=(x >= 1.96) & (x <= 4), color ='C0')

plt.axvline(x=1.96, linestyle='--')
plt.axvline(x=-1.96, linestyle='--')

plt.title('Распределения разности выборочных средних при верности нулевой гиптотезы и при верности альтернативной гипотезы')
plt.ylabel('$f(x)$ - propability density function')
plt.xlabel('Разность выборочных средних')
plt.show()
```

Что видно на графике:
1. Второе распределение разности выборочных средних смещено относительно первого распределения разности выборочных средних
2. $\mu_{приH1}$ = 3, $\mu_{приH0}$ = 0
3. Синяя область - это ошибка первого рода. Вероятность найти различия там, где их нет
4. Пунктирная линия - это критическое значение
5. Зеленая область - это ошибка второго рода. Вероятность не найти разлития там, где они есть
6. Оранжевая область - это мощность теста. Вероятность найти различия там, где они есть.

**Мощность теста** - это вероятность того, что тест найдет различия там, где они есть. Иными словами обнаружит эффект там, где он есть. Мощность еще часто называют чувствительностью теста. Чем мощнее тест, тем он предпочтительней.

Мощность теста расчитывается по формуле: $1- \beta$

Представьте, что верна нулевая гипотеза, но ваша статистика попала за пределы критических значений, тогда вы совершили ошибку первого рода. Если верна альтернативная гипотеза, но ваша статистика не попала за пределы критических значений, тогда вы совершили ошибку второго рода.

### 3. Размер выборки и MDE

Когда мы запускаем АБ тест, мы ожидаем, что наша метрика (конверсия или среднее) изменится. При этом у продукт-менеджера или маркетолога или аналитика обычно есть представление о том насколько должна измениться метрика, чтобы это нас устроило. 

Если метрика вырастем ничтожно - в этом изменении может не быть смысла.

В зависимости от размера выборки (и дисперсии) ваш АБ тест может отлавливать разный размер эффекта. Эти эффекты называют MDE (Minimum detectable effect).

К примеру сейчас конверсия состовляет 0.07, а мы хотели бы, чтобы конверсия была как минимум 0.08. Мы хотим, чтобы наш тест умел обнаруживать эффект(MDE) = 0.01.

Мы можем определить какой размер выборки нам понадобится с помощью формулы:

### 3.1 Формула размера выборок при сравнении пророрций

$$ n=\frac{(Z_{\frac{\alpha}{2}}\sqrt{2*p_1 (1-p_1)}+Z_{\beta}\sqrt{p_1(1-p_1)+p_2(1-p_2)})^2}{MDE^2} $$

где:</br>
$p_1$ - конверсия до изменений </br>
$p_2$ - конверсия после изменений. $p_2 = p_1+MDE$ </br>
$Z_{\frac{\alpha}{2}}$ - z-квантиль ошибки первого рода деленного на 2 (деление 2 из за того, что тест двухвыборочный);</br>
$Z_{\beta}$ - z-квантиль ошибки второго рода;</br>
$p_1 (1-p_1)$ - дисперсия;</br>
$p_2(1-p_2)$ -дисперсия с учетом того, что среднее (конверсия) увеличится на MDE</br>
$MDE$ - минимальный детектируемый эффект. Расчитывается как $p_2 - p_1$

**Задача**:
На исторических данных мы определили, что конверсия составляет 0.07. Мы хотим увеличить конверсию до 0.08. Какой размер выборки нам понадобится?

Мы хотим фиксировать ошибки: $\alpha$ = 0.05, $\beta$ = 0.2. Мощность = 1 - $\beta$ = 0.8<br/>

```python
a = 0.05 #уровень значимости альфа, ошибка первого рода
b = 0.2 #ошибка второго рода
p_1 = 0.10 #конверсия или среднее
p_2 = 0.08 #конверсия с учетом MDE
mde = 0.08-0.07 # MDE
var = p_1*(1-p_1) #дисперсия
var_with_mde = p_2*(1-p_2)  #дисперсия с учетом MDE
```

```python
#рассчитаем критические значения для ошибки первого рода
z_a = sts.norm.ppf(0.05/2)
print(z_a)
```

```python
#рассчитаем критические значения для ошибки второго рода
z_b = sts.norm.ppf(0.2)
print(z_b)
```

```python
n = (  z_a * (2*var)**0.5 + z_b * (var + var_with_mde)**0.5 )**2/ mde**2
print(n)
```

Размер выборки также можно расчитать здесь: https://www.evanmiller.org/ab-testing/sample-size.html

### 3.1 Формула размера выборок при cравнении средних

$$n \geq \frac{(Var_{test} + Var_{control})*{(Z_{\frac{\alpha}{2}}+Z_{\beta})}^2}{{MDE}^2}$$


где:</br>
$n$ - это размер выборки;</br>
$Var_{test}$ - это дисперсия тестовой группы;</br>
$Var_{control}$ - это дисперсия контрольной группы;</br>
$Z_{\frac{\alpha}{2}}$ - z-квантиль ошибки первого рода деленного на 2 (деление 2 из за того, что тест двухвыборочный);</br>
$Z_{\beta}$ - z-квантиль ошибки второго рода;</br>
$MDE$ - минимальный детектируемый эффект.

**Задача**: У вашего супермаркета Покупочка есть приложение для покупок онлайн. Ваши разработчики разработали новый алгоритм рекомендаций товаров, который должен увеличить средний оборот на пользователя (посетителя сайта) за определенный период времени. Эту метрику еще называют ARPU (Average Revenue Per User).

На исторических данных вы знаете, что ARPU = 750 рублей, при стандартном отклонении 200 рублей. А вы хотите повысить АRPU на 20 рублей, до 770. Рассчитайте размер выборок,

Мы хотим фиксировать ошибки: $\alpha$ = 0.05, $\beta$ = 0.2. Мощность = 1 - $\beta$ = 0.8<br/>

```python
a = 0.05 #уровень значимости альфа, ошибка первого рода
b = 0.2 #ошибка второго рода
var = 200**2
mde = 770-750 # MDE
```

```python
z_a = sts.norm.ppf(0.05/2)
print(z_a)
```

```python
z_b = sts.norm.ppf(0.2)
print(z_b)
```

```python
n =  (var + var) * (z_a + z_b)**2 /mde**2
print(n)
```
