```python
import pandas as pd
df_p = pd.read_csv("Pandas_data_for_course/Products.csv", sep=';')
df_c = pd.read_csv("Pandas_data_for_course/Customers.csv", sep=';')
df_o = pd.read_csv("Pandas_data_for_course/Orders.csv", sep=';')
df_od = pd.read_csv("Pandas_data_for_course/Order_details.csv", sep=';')
df_e = pd.read_excel("Pandas_data_for_course/Employees.xlsx")
```

```python


# Преобразование столбца 'OrderDate' в формат datetime
df_o['OrderDate'] = pd.to_datetime(df_o['OrderDate'])

# Извлечение месяца и года в формате MMYYYY
df_o['MonthYear'] = df_o['OrderDate'].dt.strftime('%m%Y')

# Подсчет количества заказов по месяцам
orders_per_month = df_o['MonthYear'].value_counts()

# Нахождение месяца с максимальным количеством заказов
max_orders_month = orders_per_month.idxmax()
print(f"Месяц с рекордным количеством заказов: {max_orders_month}")
```

```python
# merge three tables 
df_o_od = df_c.merge(df_o).merge(df_od)
df_o_od['OrderDate'] = df_o_od['OrderDate'].astype('datetime64[ns]')
df_o_od['Discount_fact']= df_o_od['UnitPrice']*df_o_od['Quantity']*df_o_od['Discount']
df_o_od['Revenue'] = df_o_od.UnitPrice * df_o_od.Quantity - df_o_od['Discount_fact']
df_o_od['year'] = df_o_od['OrderDate'].dt.year
df_o_od['month'] = df_o_od['OrderDate'].dt.month
df_o_od
```

```python
df_o_od.groupby(['EmployeeID', 'month', 'year']).Revenue.sum().sort_values(ascending = False)
```

```python
# Преобразуем OrderDate в формат datetime и выделим колонку с месяцем и годом
df_o_od['OrderDate'] = pd.to_datetime(df_o_od['OrderDate'])
df_o_od['YearMonth'] = df_o_od['OrderDate'].dt.to_period('M')

# Группируем по EmployeeID и YearMonth, суммируем выручку, сортируем и выбираем максимум
record_net_revenue = df_o_od.groupby(['EmployeeID', 'YearMonth'])['Revenue'].sum().sort_values(
    ascending=False).head(1)
record_net_revenue
```

```python
# Группировка по сотрудникам, годам и месяцам, и суммирование выручки
grouped = df_o_od.groupby(['EmployeeID', 'year', 'month'])['Revenue'].sum().reset_index()

# Поиск максимальной выручки
max_revenue = grouped['Revenue'].max()

# Найти строку, соответствующую максимальной выручке
max_revenue_row = grouped.loc[grouped['Revenue'] == max_revenue]

print(max_revenue_row)
```

```python
import pandas as pd

# Приведение OrderDate к datetime64
df_o['OrderDate'] = pd.to_datetime(df_o['OrderDate'])

# Рассчитываем выручку в df_od
df_od['Revenue'] = df_od['UnitPrice'] * df_od['Quantity'] * (1 - df_od['Discount'])

# Оставляем только два столбца: OrderID и Revenue
df_od = df_od[['OrderID', 'Revenue']]

# Присоединяем df_od к df_o
df_o_od = df_o.merge(df_od, on='OrderID')

# Добавляем столбцы month и year
df_o_od['year'] = df_o_od['OrderDate'].dt.year
df_o_od['month'] = df_o_od['OrderDate'].dt.month

# Группируем по EmployeeID, year и month, суммируем Revenue
grouped = df_o_od.groupby(['EmployeeID', 'year', 'month'])['Revenue'].sum().reset_index()

# Сортируем по убыванию выручки
grouped = grouped.sort_values(by='Revenue', ascending=False)

# Выводим результат
print(grouped)
```
