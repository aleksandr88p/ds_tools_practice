```python
# заполнение NaN значений в столбце ShippedDate
df_o['ShippedDate'] = df_o['ShippedDate'].fillna("2024-09-10 00:00:00.000")
df_o
```

```python
# replace для замены конкретного значения в столбце
df_o['ShippedDate'] = df_o['ShippedDate'].replace("1996-07-16 00:00:00.000", "notshipped")
df_o
```

```python
# df_o['ShippedDate'].where() - возвращает значения, которые удовлетворяют условию
# df_o['OrderID'].isin([11076, 11077]) - возвращает True, если значение есть в списке в данном случае все OrderID которые равны 11076 или 11077

df_o['ShippedDate'].where(df_o['OrderID'].isin([11076, 11077]))
```

```python
# replace будет применен для всех значений, которые не удовлетворяют условию df_o['OrderID'].isin([11076, 11077])
df_o.ShippedDate.where(df_o.OrderID.isin([11076, 11077]), 'Notshipped')
```

```python
# replace будет применен для всех значений которй удовлетворяют условию df_o['OrderID'].isin([11076, 11077])
df_o.ShippedDate.where(~(df_o.OrderID.isin([11076, 11077])), 'Notshipped')
```
