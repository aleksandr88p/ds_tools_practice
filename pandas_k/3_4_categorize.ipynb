### Посчитайте для каждого заказа чистую выручку. Затем помощью метода .cut() разбейте столбец с чистой выручкой на 5 бинов. Добавьте столбец с бинами в датафрейм. Далее в разрезе бинов посчитайте количество заказов. 

```python
df_od = pd.read_csv("Pandas_data_for_course/Order_details.csv", sep=';')
df_od['DiscountFact'] = df_od['Discount'] * df_od['Quantity'] * df_od['UnitPrice']
df_od['Revenue'] = df_od['Quantity'] * df_od['UnitPrice'] - df_od['DiscountFact']
df_od = df_od.groupby('OrderID').Revenue.sum().reset_index()
df_od
```

```python
df_od['bins'] = pd.cut(df_od['Revenue'],bins=5)
df_od.groupby('bins', observed=False).Revenue.count()
```

### Посчитайте для каждого заказа чистую выручку. Затем с помощью метода .qcut() разбейте столбец с чистой выручкой на 5 бинов. Добавьте столбец с бинами в датафрейм. Далее в разрезе бинов посчитайте количество заказов. Выберите диапазон первого бина.


```python
df_od = pd.read_csv("Pandas_data_for_course/Order_details.csv", sep=';')
df_od['DiscountFact'] = df_od['Discount'] * df_od['Quantity'] * df_od['UnitPrice']
df_od['Revenue'] = df_od['Quantity'] * df_od['UnitPrice'] - df_od['DiscountFact']
df_od = df_od.groupby('OrderID').Revenue.sum().reset_index()
df_od['bins'] = pd.qcut(df_od['Revenue'],q=5)
df_od.groupby('bins', observed=False).Revenue.count()
```
